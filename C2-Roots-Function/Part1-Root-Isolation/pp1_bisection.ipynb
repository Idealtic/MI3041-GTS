# Phương pháp chia đôi (Bisection Method)

# Điều kiện:

* $(a,b)$ là khoảng cách ly nghiệm

* $f(x)$ liên tục trên $(a,b)$

* $f(a) \cdot f(b) < 0$

# Thuật toán tổng quát:

1. Kiểm tra input có thỏa mãn điều kiện không

$\quad$ Chọn khoảng $[a_0,b_0]$ ban đầu sao cho $f(a_0) \cdot f(b_0) < 0$

2. Xác định $x$ theo công thức lặp: $x_{n+1} = \dfrac{a_n + b_n}{2}$

3. Xác định khoảng tiếp theo $[a_{n+1}, b_{n+1}]$:

   a. Nếu $f(a_{n}) \cdot f(x_{n+1}) < 0$, giữ $a_{n+1} = a_n$ và cập nhật $b_{n+1} = x_{n+1}$

   b. Nếu $f(b_{n}) \cdot f(x_{n+1}) < 0 \Leftrightarrow f(a_{n}) \cdot f(x_{n+1}) > 0$, cập nhật $a_{n+1} = x_{n+1}$ và giữ $b_{n+1} = b_n$

   Tạo biến dấu để kiểm tra dấu của $f(x_{n+1})$: Nếu $f(x_{n+1})>0$ trả về 1, trái lại trả về -1, lưu vào sign

      a. Nếu sign $\cdot f(x_{n+1}) < 0$, giữ $a_{n+1} = a_n$ và cập nhật $b_{n+1} = x_{n+1}$
   
      b. Nếu sign $\cdot f(x_{n+1}) > 0$, cập nhật $a_{n+1} = x_{n+1}$ và giữ $b_{n+1} = b_{n}$

4. Lặp lại các bước (2) và (3) cho đến khi:

   a. $x_n$ là nghiệm của phương trình, hoặc

   b. khoảng $[a_n, b_n]$ thỏa mãn điều kiện dừng

5. In giá trị $x_n$ (giá trị $x$ sau $n$ lần lặp)

# Áp dụng


## 1. Tiên nghiệm

Đánh giá sai số: $|x_n - x^*| \leqslant b_n - a_n = \dfrac{b-a}{2^n} < \varepsilon$

### 1.1 Thuật toán
#### Input:
- Khoảng cách ly nghiệm $(a,b)$
- Sai số $\varepsilon$
- Hàm số $f(x)$
#### Output: nghiệm gần đúng $x$
#### Thuật toán:
- Bước 1: Kiểm tra điều kiện tồn tại nghiệm:
    - Nếu $f(a) \cdot f(b) > 0$: dừng chương trình do không thoả mãn điều kiện đầu vào
    - Nếu $f(a)=0$: thu được nghiệm $x=a$, dừng chương trình
    - Nếu $f(b)=0$: thu được nghiệm $x=b$, dừng chương trình
- Bước 2: Tính số lần lặp: $n=\left[\log_{2}\left(\dfrac{b-a}{\varepsilon}\right)\right]+1$
- Bước 3: Tạo biến sign lưu dấu $f(a)$:
    - Nếu $f(a)>0$: sign $=1$
    - Nếu $f(a)<0$: sign $=-1$
- Bước 4: Khởi tạo $i=0$. Khi $i \leqslant n $ thực hiện:
    - Tính $x_{i+1}=\dfrac{a_i+b_i}{2}$
    - Nếu $f(x_{i+1})=0$: thu được nghiệm $x=x_{i+1}$, dừng chương trình
    - Nếu $\text{sign} \cdot f(x_{i+1}) > 0$: cập nhật $a_{i+1}=x_{i+1}$, giữ $b_{i+1}=b_i$
    - Nếu $\text{sign} \cdot f(x_{i+1}) < 0$: cập nhật $b_{i+1}=x_{i+1}$, giữ $a_{i+1}=a_i$
    - Cập nhật $i=i+1$
- Bước 5: Sau khi kết thúc $n$ vòng lặp, thu được nghiệm gần đúng là $x=\dfrac{a_n+b_n}{2}$. Dừng chương trình.

### 1.2 Code

### 1.2.1 Có cộng thêm sai số làm tròn (chưa cho trước $\varepsilon$)
Từ công thức sai số Tiên nghiệm, suy ra Điều kiện dừng: $n > \log_2 \left( \dfrac{b-a}{\varepsilon} \right)$

```python
def bisection_iteration_v1(f,a,b,n,digit): # digit = số chữ số sau dấu phẩy
    # Input check
    if (f(a) * f(b) >= 0):
        print("You have not assumed right a and b\n")
        return

    # Chia đôi
    x = 0
    delta = 0
    sign_f_a = sign(f(a))

    results = []
    diff = b-a
    temp_2 = 2

    for i in range(n):
        # Tìm x mới
        x_new = (a+b) / 2.0
        delta = diff / temp_2
        temp_2 *= 2

        results.append({
            'i': i,
            'a_i': a,
            'b_i': b,
            'x_(i+1)': x_new,
            'f(a_i)': f(a),
            'f(b_i)': f(b),
            'f(x_(i+1))': f(x_new),
            'delta=(b-a)/2^(i+1)': delta
        })

        # Chuẩn bị cho lần lặp tiếp
        x = x_new
        if f(x_new) == 0:
            break
        elif (f(x_new) * sign_f_a < 0):
            b = x_new
        elif (f(x_new) * sign_f_a > 0):
            a = x_new

    # In kết quả
    df_results = pd.DataFrame(results)
    print(df_results.to_string(index=False))

    if digit == None:
        print(f"Nghiệm cần tìm là: {x}")
    else:
        total_delta = delta + 0.5 * 10**(-digit) # Tính toán sai số làm tròn
        print(f"Vậy x xấp xỉ x_{n} = {x}")
        print(f"Nghiệm cần tìm với {digit} chữ số sau dấu phẩy: {round(x, digit)}")
        print(f"Sai số tuyệt đối: {total_delta}")
```

## 2. Hậu nghiệm - Sai số tuyệt đối

Đánh giá sai số: $|x_n - x^*| \leqslant b_n - a_n = |x_n - x_{n-1}|$

### 2.1 Thuật toán
#### Input:
- Khoảng cách ly nghiệm $(a,b)$
- Sai số $\varepsilon$
- Hàm số $f(x)$
#### Output: nghiệm gần đúng $x$, số lần lặp $n$
#### Thuật toán:
- Bước 1: Kiểm tra điều kiện tồn tại nghiệm:
    - Nếu $f(a) \cdot f(b) > 0$: dừng chương trình do không thoả mãn điều kiện đầu vào
    - Nếu $f(a)=0$: thu được nghiệm $x=a$, dừng chương trình
    - Nếu $f(b)=0$: thu được nghiệm $x=b$, dừng chương trình
- Bước 2: Tạo biến sign lưu dấu $f(a)$:
    - Nếu $f(a)>0$: sign $=1$
    - Nếu $f(a)<0$: sign $=-1$
- Bước 3: Khởi tạo $i=0$. Lặp đến khi thoả mãn điều kiện dừng $|x_n-x_{n-1}|< \varepsilon$ :
    - Tính $x_{i+1}=\dfrac{a_i+b_i}{2}, \quad \Delta x_{i}=|x_{i+1}-x_i|$
    - Nếu $f(x_{i+1})=0$: thu được nghiệm $x=x_{i+1}$, dừng chương trình
    - Nếu $\text{sign} \cdot f(x_{i+1}) > 0$: cập nhật $a_{i+1}=x_{i+1}$, giữ $b_{i+1}=b_i$
    - Nếu $\text{sign} \cdot f(x_{i+1}) < 0$: cập nhật $b_{i+1}=x_{i+1}$, giữ $a_{i+1}=a_i$
    - Nếu thoả mãn điều kiện dừng $\Delta x_{i} < \varepsilon$: dừng chương trình. Trái lại cập nhật $i=i+1$
- Bước 4: Sau khi thoả mãn điều kiện dừng với $n$ lần lặp, thu được nghiệm gần đúng là $x=\dfrac{a_n+b_n}{2}$. Dừng chương trình.

### 2.2 Code
### 2.2.1 Có cộng thêm sai số làm tròn (chưa cho trước $\varepsilon$)

```python
def bisection_iteration_v2 (f,a,b,n,digit):
    # Input check
    if (f(a) * f(b) >= 0):
        print("You have not assumed right a and b\n")
        return

    # Chia đôi
    x = 0
    delta_x = 0
    sign_f_a = sign(f(a))

    results = []
    for i in range(n):
        # Tìm x mới
        x_new = (a+b) / 2.0
        delta_x = abs(x_new - x)

        results.append({
            'i': i,
            'a_i': a,
            'b_i': b,
            'x_(i+1)': x_new,
            'f(a_i)': f(a),
            'f(b_i)': f(b),
            'f(x_(i+1))': f(x_new),
            'delta_x=|x_(i+1)-x_i|': delta_x
        })

        # Chuẩn bị cho lần lặp tiếp
        x =  x_new
        if f(x_new) == 0:
            break
        elif (f(x_new) * sign_f_a < 0):
            b = x_new
        elif (f(x_new) * sign_f_a > 0):
            a = x_new

    # In kết quả
    df_results = pd.DataFrame(results)
    print(df_results.to_string(index=False))

    if digit == None:
        print(f"Nghiệm cần tìm là: {x}")
    else:
        total_delta = delta_x + 0.5 * 10**(-digit) # Tính toán sai số làm tròn
        print(f"Vậy x xấp xỉ x_{n} = {x}")
        print(f"Nghiệm cần tìm với {digit} chữ số sau dấu phẩy: {round(x, digit)}")
        print(f"Sai số tuyệt đối: {total_delta}")
```

```python
f = lambda x: np.e**x - np.cos(2*x)

a = -3
b = -2

n = 30
digit = 5

bisection_iteration_v2 (f, a, b, n, digit)
```

### 2.2.2 Không cộng thêm sai số làm tròn (đã cho trước $\varepsilon$)

Từ công thức ta có: $|x_n - x^*| \leqslant |x_n - x_{n-1}| < \epsilon$ - điều kiện dừng (hậu nghiệm)

```python
def bisection_recursion_absolute (f,a,b,eps):
    # Error function
    if (f(a) * f(b) >= 0):
        print("You have not assumed right a and b\n")
        return

    # Chia đôi
    x = 0
    sign_f_a = sign(f(a))
    print(f"delta_x = {eps}")

    i = 0
    results = []
    while True:
        # Tìm x mới
        x_new = (a+b) / 2.0
        current_delta_x = abs(x_new - x)

        results.append({
            'i': i,
            'a_i': a,
            'b_i': b,
            'x_(i+1)': x_new,
            'f(a_i)': f(a),
            'f(b_i)': f(b),
            'f(x_(i+1))': f(x_new),
            'delta_x=|x_(i+1)-x_i|': current_delta_x
        })

        # Chuẩn bị cho lần lặp tiếp theo
        x = x_new
        if f(x_new) == 0:
            break
        elif (f(x_new) * sign_f_a < 0):
            b = x_new
        elif (f(x_new) * sign_f_a > 0):
            a = x_new

        # Điều kiện dừng
        if current_delta_x < eps:
            break
        else:
            i += 1

    # In kết quả
    df_results = pd.DataFrame(results)
    print(df_results.to_string(index=False))

    print(f"Giá trị cần tìm với sai số tuyệt đối {eps} là: {x}")
```

Câu 3.1. Viết thuật toán tính gần đúng hằng số $e$ với sai số $\varepsilon$ cho trước. Áp dụng tính $e$ với bảy chữ số đáng tin sau dấu phẩy.

```python
f = lambda x: np.log(x) - 1 #approximate e

a = 2
b = 3

eps = 0.5 * pow(10, -7) #epsilon should be small so that the root is accurate to 7 decimal places

bisection_recursion_absolute(f, a, b, eps)
```

### 2.2.3 Tìm nghiệm thoả mãn điều kiện sai số tương đối

Từ công thức ta có: $\dfrac{|x_n - x^*|}{|x_n|} \leqslant \dfrac{|x_n - x_{n-1}|}{|x_n|} < \eta$ - điều kiện dừng

```python
def bisection_recursion_relative (f,a,b,eta):
    # Error function
    if (f(a) * f(b) >= 0):
        print("You have not assumed right a and b\n")
        return

    # Chia đôi
    x = 0
    sign_f_a = sign(f(a))
    print(f"sigma_x = {eta}")

    i = 0
    results = []
    while True:
        # Tìm x mới
        x_new = (a+b) / 2.0
        current_sigma_x = abs(x_new - x)/abs(x_new)

        results.append({
            'i': i,
            'a_i': a,
            'b_i': b,
            'x_(i+1)': x_new,
            'f(a_i)': f(a),
            'f(b_i)': f(b),
            'f(x_(i+1))': f(x_new),
            'sigma_x=|x_(i+1)-x_n|/|x_(i+1)': current_sigma_x
        })

        # Chuẩn bị lần lặp tiếp theo
        x = x_new
        if f(x_new) == 0:
            break
        elif (f(x_new) * sign_f_a < 0):
            b = x_new
        elif (f(x_new) * sign_f_a > 0):
            a = x_new

        # Điều kiện dừng
        if current_sigma_x < eta:
            break
        else:
            i += 1

    # Print the final result
    df_results = pd.DataFrame(results)
    print(df_results.to_string(index=False))

    print(f"Giá trị cần tìm với sai số tương đối {eta} là: {x}")
```

```python
f = lambda x: np.exp(x) - np.cos(2*x)

a = -3
b = -2

eta = 0.5 * pow(10, -6)

bisection_recursion_relative (f, a, b, eta)
```
